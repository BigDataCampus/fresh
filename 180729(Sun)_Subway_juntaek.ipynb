```python
soap = []
[soap.append(BeautifulSoup(resp[i].text, 'html.parser')) for i in range(9)]
```

```python
soap[0]
```

```python
x = []
[x.append(soap[i].select('div.subway > div > dl > dd > ul > li')) for i in range(9)]
```

```python
x[0]
```

```python
subwaylist = [[] for i in range(9)]
[subwaylist[j].append(i.get_text()) for j in range(9) for i in x[j]]
```

```python
subwaylist[1]
```

```python
subwayimg = [[] for i in range(9)]
[subwayimg[j].append(i.select('img')) for j in range(9) for i in x[j]]
```

```python
subwayimg[1]
```

```python
result = [dict(zip(subwaylist[i], subwayimg[i]))  for i in range(9)]
```
